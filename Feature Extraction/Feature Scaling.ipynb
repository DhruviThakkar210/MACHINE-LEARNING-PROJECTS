# Feature Scaling


```python
import pandas as pd 
import seaborn as sns
```

```python
df = pd.read_csv('Car_data.csv')
```

```python
df.head()
```

```python
km_driven=list(df['km_driven'])
```

```python
km_driven
```

```python

sns.histplot('km_driven')
```

```python


import numpy as np
mean=np.mean(km_driven)
std=np.std(km_driven)
```

```python
mean
```

```python
std
```

```python
normalized_data=[]
for i in km_driven:
    z_score=(i-mean)/std ## z_score
    normalized_data.append(z_score)

print(normalized_data)

```

```python
sns.histplot(normalized_data)
```
